```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('countries.csv')

kenya = data[data.country == 'Kenya']
uganda = data[data.country == 'Uganda']
tanzania = data[data.country == 'Tanzania']
rwanda = data[data.country == 'Rwanda']
burundi = data[data.country == 'Burundi']

plt.plot(kenya.year, kenya.population / 10**6)
plt.plot(uganda.year, uganda.population / 10**6)
plt.plot(tanzania.year, tanzania.population / 10**6)
plt.title('East Africa Population Increase')
plt.xlabel('Years')
plt.ylabel('Population Increase')

plt.legend(['Kenya', 'Uganda', 'Tanzania'])

plt.show()
```

```python
# The Eastern Africa graph
plt.plot(kenya.year, kenya.population / 10**6)
plt.plot(uganda.year, uganda.population / 10**6)
plt.plot(tanzania.year, tanzania.population / 10**6)
plt.plot(rwanda.year, rwanda.population / 10**6)
plt.plot(burundi.year, burundi.population / 10**6)

plt.title('The Eastern Africa Community Graph of Population Since 1952')
plt.xlabel('The years recorded')
plt.ylabel('The population increase')
plt.legend(['Kenya', 'Uganda', 'Tanzania', 'Rwanda', 'Burundi'])

plt.show()
```
