```python
# But using the keyword...
students2
```

```python
students2.drop_duplicates(inplace=True)
students2 # Look, it has solved the dups problem
```

```python
# Another important arg for drop_dups is Keep: Def= True, False, First, Last
# First keeps the first dup occurence
# Last is V.V
# False, drops all of them... First create the dup variable again
students2
```

```python
students2.drop_duplicates(inplace=True, keep=False)
students2 # See, False drops all dups
```
