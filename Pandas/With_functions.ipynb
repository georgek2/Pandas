```python
import pandas as pd
```

```python
students = pd.DataFrame({
    'names':['Jane', 'Lukas', 'Joseph', 'Mark', 'Erick', 'Anna', 'Ford', 'Lucifer', 'James', 'You'],
    'English':[33, 28, 31, 32, 27, 30, 36, 42, 37, 36],
    'Science':[25, 35, 29, 37, 34, 36, 45, 47, 32, 44],
    'History':[37, 34, 40, 38, 35, 41, 39, 44, 33, 41],
    'Mathematics':[33, 34, 27, 37, 31, 27, 31, 44, 37, 42],
    'Religious':[37, 33, 35, 39, 26, 37, 36, 50, 26, 40]
})
students.set_index('names', inplace=True)
students
```

```python
# A function that creates a new col and adds the Total for each student..
def total(name):
    summed = students.loc[name].sum()
    return summed

total('Ford')
```

```python
# Determine if it has reached the needed score..
def again(x):
    if x >= 40:
        return 'Mathematician'
    else:
        return 'Not there yet'
```

```python
# Wow, amazing.. Adding a new col by using a function and determining from another col in the dataframe..
students['Mathematician'] = students['Mathematics'].apply(again)
students
```

```python
# Find a promising scientist
students['Scientist'] = students['Science'] >= 35
students
```

```python
# Create a new col to find future engineers.
# Use the Maths and Science marks above 40
students['Engineers'] = students[((students['Science'] >= 40) & (students['Mathematics'] >= 40))]
students
```

```python
# Those with 40 and above in both Sci and Hist
students['Engineer'] = students['Science'] & students['History'] >= 40
students
```

```python
# Determine future scholars
def the_det(x):
    if x >= 40:
        return 'Scholar'
    else:
        return 'Not Scholar'
```

```python
students['Scholar'] = students['Mathematics'].apply(the_det)
students
```

```python
# Trying to calc the totals and add the col for it using a function 
students
```

```python
# Calc total function
def calc_total():
    student_mark = students.iloc[0:]
    total = student_mark.sum()
    return total

# calc_total() # The function works. The axis=0 though

# We want summations of the cols
def calc_sum():
    total = students.sum(axis=1) # Setting the axis equal to 1
    return total

calc_sum()
```

```python
# Find out who's more Godly using their religious subj mark
def the_holier(x):
    if x >= 37:
        return 'Godly'
    else:
        return 'Not tha much'
# Can and should be replaced by the lambda function below which achieves the same effect...
```

```python
students['Who\'s Godly'] = students['Religious'].apply(the_holier)
students
```

```python
# The lambda/ Anonymous function
students['F_Engineers'] = students['Science'].apply(lambda x: 'Promising' if x >= 35 else 'None')
students # See, one liner and the exact same effect
```
